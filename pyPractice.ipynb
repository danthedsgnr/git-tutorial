```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Prepare the Dataset
data = pd.DataFrame({
    'Weight': [150, 180, 170, 200],
    'Texture': ['Smooth', 'Smooth', 'Rough', 'Rough'],
    'Label': ['Apple', 'Apple', 'Orange', 'Orange']
})

# Encode categorical data
data['Texture'] = data['Texture'].map({'Smooth': 0, 'Rough': 1})  # Smooth = 0, Rough = 1
data['Label'] = data['Label'].map({'Apple': 0, 'Orange': 1})      # Apple = 0, Orange = 1

# Features (X) and target (y)
X = data[['Weight', 'Texture']]
y = data['Label']

# Step 2: Train Decision Tree Classifier
model = DecisionTreeClassifier()
model.fit(X, y)

# Step 3: Visualize the Decision Tree
plt.figure(figsize=(10, 6))
plot_tree(model, feature_names=['Weight', 'Texture'], class_names=['Apple', 'Orange'], filled=True)
plt.show()

# Step 4: Make Predictions
sample = np.array([[160, 0]])  # Example: Weight=160, Texture=Smooth
prediction = model.predict(sample)
print(f"Prediction (0=Apple, 1=Orange): {prediction[0]}")
```
